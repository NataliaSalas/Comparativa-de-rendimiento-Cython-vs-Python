#**Introducción**

Programar es un proceso que consiste en crear una secuencia de algoritmos y un fichero
donde se ponen estas instrucciones utilizando un lenguaje de programación para que
posteriormente, pueda ser ejecutado por una máquina. Los programas son un conjunto de
sentencias escritas donde un lenguaje de programación le dice al computador que tareas debe
realizar [1]. 

Existen dos tipos de lenguajes de programación, compilado e interpretado. C es un lenguaje compilado, esto quiere decir que se traduce el código fuente
escrito a código máquina empleando compiladores, generándose de esta manera el programa
ejecutable por el computador. Es un lenguaje de programación de alto nivel,  de propósito general, lo que significa que
se puede utilizar para hacer programas de todo tipo, desde un juego hasta un sistema
operativo. Tiene alta transportabilidad y es compacto, ya que solo tiene 32 palabras
reservadas. Además, es un lenguaje estructurado, puesto que los programas
escritos con él se pueden organizar en módulos [2].

Python es un lenguaje de programación interpretado, que es utilizado para varios propósitos, entre ellos aplicaciones web, ciencias de datos y machine learning. Este es uno de los lenguajes de programación más utilizados debido a su eficiencia y a que es fácil de aprender. También a su característica multiplataforma, la integración de todos los tipos de sistemas que aumenta la velocidad de desarrollo.[3]

Una de las formas en que se puede acelerar Python es utilizando Cython. Este es un lenguaje de programación que fusiona lo mejor de dos mundos. Combinado la eficiencia y fácil manejo de Python con el sistema de tipos estáticos y el alto rendimiento de C y
C ++. Este traduce el código fuente de Cython a un código fuente de C y se puede compilar en un módulo de extensión de Python o en un ejecutable
independiente, para así obtener de esta forma considerables mejoras del rendimiento.

#**Resultados**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##*Primer Algoritmo*
El primer algoritmo que se implemento para realizar el analisis de la comparativa de rendimiento cython/python es el algoritmo distance levenchtein. 

```python
levenshtein1 = pd.read_csv("/content/levenshtein1.csv")
levenshtein2 = pd.read_csv("/content/levenshtein2.csv")
```

```python
levenshtein1.head()
```

```python
levenshtein2.head()
```

```python
levenshtein1_py=levenshtein1['py'].mean()
levenshtein2_py=levenshtein1['py'].mean()
print('Promedio del tiempo de ejecución de la primera versión en python: ',levenshtein1_py)
print('Promedio del tiempo de ejecución de la segunda versión en python: ',levenshtein2_py)
```

```python
levenshtein1_cy=levenshtein1['cy'].mean()
levenshtein2_cy=levenshtein2['cy'].mean()
print('Promedio del tiempo de ejecución de la primera versión en cython: ',levenshtein1_cy)
print('Promedio del tiempo de ejecución de la segunda versión en cython: ',levenshtein2_cy)
```

```python
levenshtein1_mean=levenshtein1['py/cy'].mean()
levenshtein2_mean=levenshtein2['py/cy'].mean()
print('Promedio del tiempo de ejecución de la primera versión de py/cy: ',levenshtein1_mean)
print('Promedio del tiempo de ejecución de la primera versión de py/cy: ',levenshtein2_mean)
```

```python
plt.figure(figsize=(8,5))
x_axis = np.arange(len([levenshtein1_py,levenshtein2_py]))
plt.bar(x_axis + 0.2,[levenshtein1_py,levenshtein2_py], 0.4, label = 'Python', alpha=0.8)
plt.bar(x_axis - 0.2,[levenshtein1_cy,levenshtein2_cy], 0.4, label = 'Cython', alpha=0.8)
plt.xticks(x_axis,['Primera Versión','Segunda Versión'])
plt.xlabel('Versiones')
plt.ylabel('Mean Time (s)')
plt.title('Versiones Vs Tiempo')
plt.suptitle('Algoritmo Distance Lenevenshtein')
plt.yticks(np.arange(0, 0.0041, 0.0005))
plt.legend()
def addlabels(x, y, val):
    for i in range(len(x)):
        plt.text(i + val , y[i] + 0.00002, round(y[i],5), ha = 'center')

addlabels(['Primera Versión','Segunda Versión'], [levenshtein1_py,levenshtein2_py], + 0.2)
addlabels(['Primera Versión','Segunda Versión'], [levenshtein1_cy,levenshtein2_cy], - 0.2)
plt.show()
```

```python
plt.plot([1,2],[levenshtein1_py,levenshtein2_py], label = 'Python', alpha=0.8)
plt.scatter([1,2],[levenshtein1_py,levenshtein2_py])

plt.plot([1,2],[levenshtein1_cy,levenshtein2_cy],label = 'Cython', alpha=0.8)
plt.scatter([1,2],[levenshtein1_cy,levenshtein2_cy])

plt.xlabel('Versiones')
plt.ylabel('Mean Time (s)')
plt.suptitle('Algoritmo Distance Lenevenshtein')
plt.title('Versiones Vs Tiempo')
plt.legend()
plt.show()
```
